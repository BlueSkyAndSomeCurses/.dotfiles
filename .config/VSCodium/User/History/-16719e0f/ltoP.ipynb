```python
incomplete_fam_mean = data_f.loc[data_f.Pstatus =="A"].Dalc.mean() + data_f.loc[data_f.Pstatus ==  "A"].Walc.mean()
complete_fam_mean = data_f.loc[data_f.Pstatus =="T"].Dalc.mean() + data_f.loc[data_f.Pstatus ==  "T"].Walc.mean()
orphans_cons_mean = data_f.loc[data_f.guardian == "other"].Dalc.mean() + data_f.loc[data_f.guardian == "other"].Walc.mean()
print(f"Consumption in complete families: {complete_fam_mean}")
print(f"Consumption in incomplete families: {incomplete_fam_mean}")
print(f"Consumption of alcohol between orphans: {orphans_cons_mean}")
```

```python
complete_fams = pd.DataFrame()
complete_fams["alcohol"] = data_f.loc[data_f.Pstatus =="T"].Dalc + data_f.loc[data_f.Pstatus =="T"].Walc 
complete_fams["grades"] = data_f.loc[data_f.Pstatus =="T"].G3

complete_fams = complete_fams.groupby("alcohol").grades.mean()

plt.plot(complete_fams.index.tolist(), complete_fams.values.tolist())
plt.xlabel('alcohol consumption')
plt.ylabel('mean grade')
plt.show()
```

```python
alc_stud_health = data_f.loc[: , ["Dalc", "Walc", "health", "studytime"]]
alc_stud_health["alcohol"] = alc_stud_health.Dalc + alc_stud_health.Walc
alc_stud_health.drop(["Dalc", "Walc"], axis = 1, inplace = True)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(alc_stud_health.alcohol, alc_stud_health.studytime, alc_stud_health.health, c=alc_stud_health.health, cmap = "plasma", marker='o')

# Add labels and title
ax.set_xlabel("alcohol consumption")
ax.set_ylabel("study time")
ax.set_zlabel("health")
ax.set_title('3D Scatter Plot Example')

# Show the plot
plt.show()

print("Corelation between health and alcohol consumption",alc_stud_health.groupby("health").alcohol.mean())
print("Corelation between health and studyhours",alc_stud_health.groupby("health").studytime.mean())
```

### Who drinks more, male or female students?

```python
fems = data_f[data_f['sex'] == 'F']
mls = data_f[data_f['sex'] == 'M']
fem_sum_d = fems['Dalc'].sum()
fem_sum_w = fems['Walc'].sum()
sum_fem = fem_sum_w+fem_sum_d
rows_fems = fems.shape[0]
average_f=sum_fem/rows_fems
mls_sum_d = mls['Dalc'].sum()
mls_sum_w = mls['Walc'].sum()
sum_mls = mls_sum_w+mls_sum_d
rows_mls = mls.shape[0]
average_m=sum_mls/rows_mls
if average_f>average_m:
    print(f'Girls drink more alcohol. Girls - {average_f.round(2)}/5, \
boys - {average_m.round(2)}/5.')
if average_f<average_m:
    print(f'Boys drink more alcohol. Girls - {average_f.round(2)}/5, \
boys - {average_m.round(2)}/5.')
else: 
    print(f'Boys and girls drink equally({average_f}/5)')


```

### Corelation between studies and absences 

```python
studytime4 = data_f[data_f['studytime'] == 4]
studytime3 = data_f[data_f['studytime'] == 3]
studytime2 = data_f[data_f['studytime'] == 2]
studytime1 = data_f[data_f['studytime'] == 1]
average4 = studytime4['absences'].mean()
average3 = studytime3['absences'].mean()
average2 = studytime2['absences'].mean()
average1 = studytime1['absences'].mean()
if average4<=average3<=average2<=average1 or average4>=average3>=average2>=average1:
    print('There is a connection between class attendance and time spent studying.')
else:
    print('There is no connection between class attendance and time spent studying.')
```
